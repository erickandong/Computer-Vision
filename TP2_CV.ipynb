```python
import cv2
from IPython.display import Image, display

# Charger le modèle pré-entraîné pour la détection de visages
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Charger l'image sur laquelle nous allons évaluer la détection de visages
image = cv2.imread('/content/drive/MyDrive/Vision_par_ordinateur/Face_Detection/Image1.jpeg')

# Convertir l'image en niveaux de gris (la méthode de Viola-Jones fonctionne mieux sur les images en niveaux de gris)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Définir la taille minimale et maximale des visages que nous voulons détecter (en pixels)
min_face_size = (30, 30)
max_face_size = (500, 500)

# Définir le pas de déplacement (overlap) entre les fenêtres (en pourcentage)
overlap_percent = 0.02 # 50% de recouvrement

# Parcourir différentes tailles de fenêtres
for window_scale in [1.0, 0.8, 0.6]:
    # Appliquer le changement d'échelle à la taille de la fenêtre
    window_size = (int(min_face_size[0] * window_scale), int(min_face_size[1] * window_scale))

    # Appliquer la technique de fenêtre glissante avec un certain pas de déplacement (overlap)
    for overlap in [0.1, 0.3, 0.5]:
        # Convertir le pourcentage de recouvrement en pixels
        overlap_pixels = int(window_size[0] * overlap)

        # Détecter les visages avec la méthode de Viola-Jones en utilisant la fenêtre glissante
        faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=window_size, maxSize=max_face_size, flags=cv2.CASCADE_SCALE_IMAGE)

        # Dessiner des rectangles autour des visages détectés
        for (x, y, w, h) in faces:
            cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Enregistrer l'image avec les rectangles des visages détectés
        output_image_path = f"output_{window_size}_{overlap_percent}.jpg"
        cv2.imwrite(output_image_path, image)

        # Afficher le nombre de visages détectés et les paramètres utilisés
        print(f"Nombre de visages détectés avec une fenêtre de taille {window_size} et un overlap de {overlap_percent*100}%: {len(faces)}")

        # Afficher l'image avec les rectangles des visages détectés
        display(Image(output_image_path))
```

```python
import cv2
from IPython.display import Image, display

# Charger le modèle pré-entraîné pour la détection de visages
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Charger l'image sur laquelle nous allons évaluer la détection de visages
image = cv2.imread('/content/drive/MyDrive/Vision_par_ordinateur/Face_Detection/Image2.jpeg')

# Convertir l'image en niveaux de gris (la méthode de Viola-Jones fonctionne mieux sur les images en niveaux de gris)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Appliquer une égalisation d'histogramme pour améliorer la qualité de l'image
gray_image = cv2.equalizeHist(gray_image)

# Définir les paramètres de détection
scaleFactor = 1.0015
minNeighbors = 5
min_face_size = (20, 20)
max_face_size = (300, 300)

# Détecter les visages avec la méthode de Viola-Jones
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=minNeighbors, minSize=min_face_size, maxSize=max_face_size)

# Dessiner des rectangles autour des visages détectés
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Enregistrer l'image avec les rectangles des visages détectés
output_image_path = "output_image.jpg"
cv2.imwrite(output_image_path, image)

# Afficher l'image avec les rectangles des visages détectés
display(Image(output_image_path))

```

```python
from transformers import DetrImageProcessor, DetrForObjectDetection
import torch
from PIL import Image, ImageDraw, ImageFont
import os

# Charger l'image à partir du fichier
file = "/content/drive/MyDrive/Vision_par_ordinateur/Voiture_Detection/Voiture4.jpeg"
image = Image.open(file)

# Initialisation du processeur et du modèle
processor = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50", revision="no_timm")
model = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50", revision="no_timm")

# Préparation des entrées et obtention des sorties du modèle
inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)

# Convertir les sorties en format COCO API et filtrer par un score minimum
target_sizes = torch.tensor([image.size[::-1]])
results = processor.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=0.9)[0]

# Dessiner les boîtes englobantes sur l'image uniquement pour les voitures
draw = ImageDraw.Draw(image)
font = ImageFont.load_default()  # Chargement d'une police par défaut

# ID de l'étiquette associée aux voitures (à remplacer par l'ID réel)
car_label_id = 3

for _, label, box in zip(results["scores"], results["labels"], results["boxes"]):
    if label.item() == car_label_id:
        box = [round(i, 2) for i in box.tolist()]
        draw.rectangle(box, outline='red', width=3)

# Enregistrer l'image modifiée dans le répertoire "ErickaPNG"
output_dir = "/content/drive/MyDrive/Vision_par_ordinateur/ErickaPNG"
os.makedirs(output_dir, exist_ok=True)
output_file = os.path.join(output_dir, "modified_image.jpg")
image.save(output_file)

# Afficher le chemin d'accès au fichier enregistré
print(f"Image modifiée enregistrée sous : {output_file}")

# Afficher l'image modifiée
image.show()
```
